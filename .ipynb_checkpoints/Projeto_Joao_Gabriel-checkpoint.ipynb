# Introdução 

Sistemas caóticos aparecem o tempo todo em Física, desde séculos atrás, quando não se conhecia uma teoria para problemas deste tipo, até os dias de hoje. Tais sistemas apresentam uma alta sensibilidade com respeito as condições iniciais, além de não ser possível fazer previsões a longo prazo. Para contornar esta falta de precisão (quanto as previsões desses sistemas) tendo em vista a dificuldade de uma resolução analítica, usa-se, a fim de estudar a dinâmica dos mesmos a partir de computadores, métodos numéricos implementados.


Um dos sistemas caóticos mais conhecidos é o pêndulo duplo. Um pêndulo duplo é uma montagem que envolve uma haste com uma das extremidades presa à um ponto de suspensão, em sua outra extremidade há uma massa na qual é ligada outra haste portando um corpo massivo.

![](http://upload.wikimedia.org/wikipedia/commons/4/45/Double-compound-pendulum.gif)

Neste trabalho será investigado e discutido a dinâmica do pêndulo duplo convencional e submetido a algumas modificações do mesmo.

A validação do método numérico será feita através da solução de uma equação diferencial de segunda ordem com solução já conhecida.

# Equações de Movimento

Inicialmente considere que no eixo no qual está suspenso o pêndulo duplo a energia potencial é nula. Um sistema de coordenadas é definado onde este eixo coincide com o eixo-x e o eixo-y é definido como um eixo ortogonal ao eixo-x e que aponta para baixo, ou seja, é crescente para baixo, como mostrado na Figura 1.

<a name="P2s"></a>
<img src="Imagens/P2s.jpeg" width=50%; align="center" title="Pênduloduplo" />
<figcaption style="text-size:18pt"><center><b>Figura 1:</b> Pêndulo Duplo. </center></figcaption>
<p style ='text-align:justify;'>

A partir da Figura 1 note que:

\begin{equation}
x_{1} = l_{1} sen(\theta_{1})\\
y_{1} = l_{1}cos(\theta_{1})\\
x_{2} = l_{1} sen(\theta_{1}) + l_{2} sen(\theta_{2})\\
y_{2} = l_{1}cos(\theta_{1}) + l_{2}cos(\theta_{2})
\end{equation}

Partindo destas equações, pode-se expressar as energias cinética e potencial do sistema fazendo:

\begin{equation}
T = \frac{1}{2} m_{1}(\dot{x_{1}}^2 + \dot{y_{1}}^2) + \frac{1}{2} m_{2}(\dot{x_{2}}^2 + \dot{y_{2}}^2)\\
V = - m_{1}gy_{1} - m_{2}gy_{2}
\end{equation}

onde,

\begin{equation}
\dot{x_{1}}^2 = l_{1}^2 \dot{\theta_{1}}^2cos^2(\theta_{1}) \\
\dot{y_{1}}^2 = l_{1}^2 \dot{\theta_{1}}^2sen^2(\theta_{1})\\
\dot{x_{2}}^2 = l_{1}^2 \dot{\theta_{1}}^2cos^2(\theta_{1}) + l_{2}^2 \dot{\theta_{2}}^2cos^2(\theta_{2}) +2l_{1} l_{2}\dot{\theta_{1}}\dot{\theta_{2}}cos(\theta_{1})cos(\theta_{2})\\
\dot{y_{2}}^2 = l_{1}^2 \dot{\theta_{1}}^2sen^2(\theta_{1}) + l_{2}^2 \dot{\theta_{2}}^2sen^2(\theta_{2}) +2l_{1} l_{2}\dot{\theta_{1}}\dot{\theta_{2}}sen(\theta_{1})sen(\theta_{2})
\end{equation}

de modo que

\begin{equation}
\dot{x_{1}}^2 + \dot{y_{1}}^2 = l_{1}^2\dot{\theta_{1}}^2\\
\dot{x_{2}}^2 + \dot{y_{2}}^2 = l_{1}^2\dot{\theta_{1}}^2 + l_{2}^2\dot{\theta_{2}}^2 + 2l_{1}l_{2}\dot{\theta_{1}}\dot{\theta_{2}}cos(\theta_{1} - \theta_{2})
\end{equation}

Assim pode-se concluir que a Lagrangeana ($L = T -V$) do sistema é:

\begin{equation}
L = \frac{1}{2} (m_{1} + m_{2})l_{1}^2\dot{\theta_{1}}^2 + \frac{1}{2} m_{2}l_{2}^2\dot{\theta_{2}}^2 + m_{2}l_{1} l_{2}\dot{\theta_{1}}\dot{\theta_{2}}cos(\theta_{1} - \theta_{2}) + (m_{1} + m_{2})gl_{1}cos(\theta_{1}) + m_{2}gl_{2}cos(\theta_{2})
\end{equation}

As equações de movimento são encontradas a partir de:

\begin{equation}
\frac{d}{dt}(\frac{\partial L}{\partial \dot{q_{k}}}) - \frac{\partial L}{\partial q_{k}} = 0, k = 1, ..., n.
\end{equation} 

Pode-se verificar que as equações de movimento são:

\begin{equation}
\ddot{\theta_{1}} + \frac{m_{2}}{m_{1} + m_{2}}\frac{l_{2}}{l_{1}}[\ddot{\theta_{2}}cos(\theta_{1} - \theta_{2}) + \dot{\theta_{2}}^2sen(\theta_{1} - \theta_{2})] + \frac{g}{l_{1}}sen(\theta_{1}) = 0 \\
\end{equation}

\begin{equation}
\ddot{\theta_{2}} + \frac{l_{1}}{l_{2}}[\ddot{\theta_{1}}cos(\theta_{1} - \theta_{2}) - \dot{\theta_{1}}^2sen(\theta_{1} - \theta_{2})]+ \frac{g}{l_{2}}sen(\theta_{2}) = 0
\end{equation}

Para resolver estas equações diferenciais de segunda ordem não-lineares e acopladas será implementado o método de Runge-Kutta de Quarta Ordem.

# Método de Runge-Kutta de Quarta Ordem (Método RK4)

Os métodos de Runge–Kutta são uma família importante de metódos iterativos implícitos e explícitos para a resolução numérica (aproximação) de soluções de equações diferenciais ordinárias (EDO). 
A família destes métodos iterativos de resolução numérica foram desenvolvidas pelos matemáticos C. Runge e M.W. Kutta no final do séc. XIX e início do séc. XX.

O método clássico de RK4 para a EDO de segunda ordem
\begin{equation}
   \ddot{y}(x) = \frac{d^{2}y}{dt^{2}} = f(t,y,\dot{y})
\end{equation} 

\begin{align*}
\dot{y}(t_{0}) = \dot{y}_{0}\\
y(t_{0}) = y_{0}
\end{align*}

é definido de acordo com o sistema abaixo:

\begin{equation}
   \dot{y}_{n+1} = \dot{y}_{n} + \frac{1}{6}(k_{1} + 2k_{2} + 2k_{3} + k_{4}) \\
\end{equation}

\begin{equation}
   y_{n+1} = y_{n} + h \cdot [\dot{y}_{n} +\frac{1}{6}(k_{1} + k_{2} + k_{3})] + O(h^{5})
\end{equation} 

onde
\begin{align*}
k_{1} &= h \cdot f(t_{n}, y_{n},\dot{y}_{n})\\
k_{2} &= h \cdot f(t_{n} +\frac{h}{2}, y_{n} + \frac{h}{2}\dot{y}_{n}, \dot{y}_{n} + \frac{k_{1}}{2})\\
k_{3} &= h \cdot f(t_{n} +\frac{h}{2}, y_{n} + \frac{h}{2}\dot{y}_{n} +\frac{h}{4}k_{1} , \dot{y}_{n} + \frac{k_{2}}{2})\\
k_{4} &= h \cdot f(t_{n} + h, y_{n} + h\dot{y}_{n} +\frac{h}{2}k_{2}, \dot{y}_{n} + k_{3})\\
\end{align*}

# Modelos Derivados do Pêndulo Duplo

Nesta seção serão indicadas as diferentes situações fazendo algumas modificações no sistema apresentado anteriormente.

A primeira modificação é um pêndulo duplo em que é trocado a primeira haste por uma mola de constante elástica $k_{1}$.

<a name="P2k1"></a>
<img src="Imagens/P2k1.jpeg" width=50%; align="center" title="Pênduloduplok1" />
<figcaption style="text-size:18pt"><center><b>Figura 3:</b> Pêndulo Duplo onde a primeira haste foi trocada por uma mola. </center></figcaption>
<p style ='text-align:justify;'>
Com isso a Lagrangeana se torna:

\begin{align}
L = \frac{1}{2}(m_{1} + m_{2})[\dot{r_{1}} + (r_{01} + r_{1})\theta_{1}^{2}] + \frac{1}{2} m_{2}l_{2}\theta_{2}^{2} + m_{2}l_{2}\dot{r_{1}}\dot{\theta_{2}}sen(\theta_{1} -\theta_{2}) + m_{2}l_{2}(r_{01} + r_{1})\dot{\theta_{1}}\dot{\theta_{2}}sen(\theta_{1} -\theta_{2}) + (m_{1} + m_{2})g(r_{01} + r_{1})cos(\theta_{1}) + m_{2}gl_{2}cos(\theta_{2}) - \frac{1}{2}k_{1}(r_{01} + r_{1} - l)^{2}
\end{align}

e as equações de movimento agora são as seguintes:

\begin{align}
\ddot{r_{1}} + \frac{m_{2}}{m_{1} + m_{2}}l_{2}[\ddot{\theta_{2}}sen(\theta_{1} - \theta_{2}) - \dot{\theta_{2}}^2 cos(\theta_{1} - \theta_{2})] - (r_{01} + r_{1})\dot{\theta_{1}}^2 - gcos(\theta_{1}) + \frac{k_{1}}{m_{1} + m_{2}}(r_{01} + r_{1} - l) = 0 \\
\end{align}

\begin{align}
\ddot{\theta_{1}} + \frac{2}{r_{01} + r_{1}} \dot{r_{1}}\dot{\theta_{1}} + \frac{m_{2}}{m_{1} + m_{2}} \frac{l_{2}}{r_{01} + r_{1}}[\ddot{\theta_{2}}cos(\theta_{1} - \theta_{2}) + \dot{\theta_{2}}^2 sen(\theta_{1} - \theta{2})] + \frac{g}{r_{01} + r_{1}}sen(\theta_{1}) = 0 \\
\end{align}

\begin{align}
\ddot{\theta_{2}} + \frac{1}{l_{2}}\ddot{r_{1}}sen(\theta_{1} - \theta_{2}) + \frac{2}{l_{2}} \dot{r_{1}}\dot{\theta_{1}}cos(\theta_{1} - \theta_{2}) + \frac{r_{01}+r_{1}}{l_{2}} [\ddot{\theta_{1}}cos(\theta_{1} - \theta_{2}) - \dot{\theta_{1}}^2 sen(\theta_{1} - \theta{2})] + \frac{g}{l_{2}}sen(\theta_{2}) = 0 
\end{align}
    
A segunda mudança no sistema é substituir a segunda haste por uma mola de constante elástica $k_{2}$ 

<a name="P2k2"></a>
<img src="Imagens/P2k2.jpeg" width=50%; align="center" title="Pênduloduplok2" />
<figcaption style="text-size:18pt"><center><b>Figura 4:</b> Pêndulo Duplo onde a segunda haste foi trocada por uma mola. </center></figcaption>
<p style ='text-align:justify;'>

desta forma a Lagrangeana é:

\begin{align} 
L = \frac{1}{2}(m_{1} + m_{2}) l_{1}^2 \dot{\theta_{1}}^2 + \frac{1}{2} m_{2} \dot{r_{2}}^2 + \frac{1}{2} m_{2} (r_{02} + r_{2}) \dot{\theta_{2}}^2 + m_{2}(r_{02} + r_{2}) l_{1}\dot{\theta_{1}}\dot{\theta_{2}}cos(\theta_{1} - \theta_{2}) - m_{2}l_{1}\dot{r_{2}}\dot{\theta_{1}}sen(\theta_{1} - \theta_{ 2}) + (m_{1} + m_{2}) l_{1}gcos(\theta_{1}) + m_{2}g(r_{02} + r_{2}cos(\theta_{2}) - \frac{k_{2}}{2}(r_{02} + r_{2} - l)^2 \\
\end{align}

Assim, as equações de movimento são escritas como:

\begin{align}
\ddot{\theta_{1}} + \frac{m_{2}}{m_{1} + m_{2}}\frac{r_{02} + r_{2}}{l_{1} }[\ddot{\theta_{2}}cos(\theta_{1} - \theta_{2}) + \dot{\theta_{2}}^2 sen(\theta_{1} - \theta_{2})] - \frac{m_{2}}{m_{1} + m_{2}} \frac{1}{l_{1}}[\ddot{r_{2}}sen(\theta_{1} - \theta_{2}) + \dot{r_{2}}\dot{\theta_{2}}cos(\theta_{1} - \theta_{2})] + \frac{g}{l_{1}}sen(\theta_{1}) = 0 \\
\end{align}

\begin{align}
\ddot{\theta_{2}} + \frac{l_{1}}{r_{02} + r_{2}}[\ddot{\theta_{1}}cos(\theta_{1} - \theta_{2}) - \dot{\theta_{1}}^2 sen(\theta_{1} - \theta_{2})] + \frac{g}{r_{02} + r_{2}}sen(\theta_{1}) = 0 \\
\end{align}

\begin{align}
\ddot{r_{2}} - l_{1}[\ddot{\theta_{1}}sen(\theta_{1} - \theta_{2}) + \dot{\theta_{1}}^2 cos(\theta_{1} - \theta_{2})] - (r_{02} + r_{2})\dot{\theta_{2}}^2 - gcos(\theta_{2}) + \frac{k_{2}}{m_{2}} (r_{02} + r_{2} - l) = 0 
\end{align}

Por fim, a última modificação é substituir as duas hastes por duas molas de constantes elásticas $k_{1}$ e $k_{2}$, respectivamente. 

<a name="P2k1k2"></a>
<img src="Imagens/P2k1k2.jpeg" width=50%; align="center" title="Pênduloduplok1k2" />
<figcaption style="text-size:18pt"><center><b>Figura 5:</b> Pêndulo Duplo onde as hastes foram trocadas por molas. </center></figcaption>
<p style ='text-align:justify;'>

A Lagrangeana que descreve o sistema agora será:

\begin{align} 
L = \frac{1}{2}(m_{1} + m_{2})[\dot{r_{1}}^2 + (r_{01} + r_{1})^2 \dot{\theta_{1}}^2] + \frac{1}{2}m_{2}[\dot{r_{2}}^2 + (r_{02} + r_{2})^2 \dot{\theta_{2}}^2] + m_{2}\dot{r_{1}}\dot{r_{2}}cos(\theta_{1} + \theta_{2}) + m_{2}(r_{01} + r_{1})(r_{02} + r_{2})\dot{\theta_{1}}\dot{\theta_{2}}cos(\theta_{1} - \theta_{2}) - m_{2}(r_{01} - r_{1})\dot{\theta_{1}}\dot{r_{2}}sen(\theta_{1} - \theta_{2}) + m_{2}(r_{02} - r_{2})\dot{\theta_{2}}\dot{r_{1}}sen(\theta_{1} - \theta_{2}) + (m_{1} + m_{2})g(r_{01} - r_{1})cos(\theta_{1}) + m_{2}g(r_{02} - r_{2})cos(\theta_{2}) - \frac{k_{1} + k_{2}}{2}(r_{01} - r_{1} - L_1)^2 - \frac{k_{2}}{2}\{(r_{02} - r_{2} - L_2)[(r_{02} - r_{2} - L_2) + 2(r_{01} - r_{1} - L_1)]\}
\end{align}

As equações correspondentes são:

\begin{align}
\ddot{r_{1}} + \frac{m_{2}}{m_{1} + m_{2}}\ddot{r_{2}}cos(\theta_{1} + \theta_{2}) + \frac{m_{2}}{m_{1} + m_{2}}(r_{02} + r_{2})[\ddot{\theta_{2}}sen(\theta_{1} - \theta_{2}) - \dot{\theta_{2}}^2 cos(\theta_{1} - \theta_{2})] - (r_{01} + r_{1})\dot{\theta_{1}}^2 + gcos(\theta_{1}) - \frac{2m_{2}}{m_{1} + m_{2}} \dot{r_{2}}(\dot{\theta_{1}} + \dot{\theta_{2}})cos(\theta_{1})sen(\theta_{2}) + \frac{(k_{1} + k_{2})}{m_{1} + m_{2}}(r_{01} + r_{1} - L_1) + \frac{k_{2}}{m_{1} + m_{2}}(r_{02} + r_{2} - L_{2}) = 0 \\
\end{align}

\begin{align}
\ddot{r_{2}} + \ddot{r_{1}}cos(\theta_{1} + \theta_{2}) - (r_{01} + r_{1})[\ddot{\theta_{1}}sen(\theta_{1} - \theta_{2}) - \dot{\theta_{1}}^2 cos(\theta_{1} - \theta_{2})] - (r_{02} + r_{2})\dot{\theta_{2}}^2 + gcos(\theta_{2}) - 2\dot{r_{1}}(\dot{\theta_{1}} + \dot{\theta_{2}})sen(\theta_{1})cos(\theta_{2} +k_{2}\{(r_{02}+r_{2} - L_{2})[1 + (r_{01} + r_{1} - L_{1})]\}= 0 \\
\end{align}

\begin{align}
\ddot{\theta_{1}} + \frac{m_{2}}{m_{1} + m_{2}}\frac{r_{02} + r_{2}}{r_{01} + r_{1}}[\ddot{\theta_{2}}cos(\theta_{1} - \theta_{2}) +\dot{\theta_{2}}^{2}sen(\theta_{1} - \theta_{2})] - \frac{m_{2}}{m_{1} + m_{2}}\frac{1}{r_{01} + r_{1}}\ddot{r_{2}}sen(\theta_{1} - \theta_{2}) + \frac{m_{2}}{m_{1} + m_{2}}\frac{1}{r_{01} + r_{1}}\dot{r_{2}}\dot{\theta_{2}}cos(\theta_{1} - \theta_{2}) + + \frac{m_{2}}{m_{1} + m_{2}}\frac{1}{r_{01} + r_{1}}\dot{r_{1}}\dot{\theta_{1}}cos(\theta_{1} - \theta_{2}) + \frac{m_{2}}{m_{1} + m_{2}}\frac{2}{(r_{01} + r_{1})^2}\dot{r_{1}}\dot{r_{2}}cos(\theta_{1})sen(\theta_{2}) + \frac{g}{r_{01} + r_{1}}sen(\theta_{1}) = 0 \\
\end{align}

\begin{align}
\ddot{\theta_{2}} + \frac{r_{01} + r_{1}}{r_{02} + r_{2}}[\ddot{\theta_{1}}cos(\theta_{1} - \theta_{2}) - \dot{\theta_{1}}^2 sen(\theta_{1} - \theta_{2})] - \frac{1}{r_{02} + r_{2}}\ddot{r_{1}}sen(\theta_{1} - \theta_{2}) + \frac{2}{r_{02} + r_{2}}\dot{r_{1}}\dot{\theta_{1}}cos(\theta_{1} - \theta_{2}) +  \frac{2}{(r_{02} + r_{2})^2}\dot{r_{1}}\dot{r_{2}}sen(\theta_{1})cos(\theta_{2}) + + \frac{g}{r_{02} + r_{2}}sen(\theta_{2}) = 0
\end{align}

Essas, juntamente com as equações de movimento da seção "Equações de movimento", são as equações as quais o programa a ser desenvolvido será capaz de resolver. Caso necessário, a implementação de um outro método este será feito no decorrer da próxima fase do projeto.

Abaixo serão escritos os módulos necessários para resolver o problema do pêndulo duplo. Os dados extraídos da solução númerica estarão sujeitos à verificação comparando com um resultado conhecido.

```python
%%writefile src/precisao.f95
MODULE precisao
!------------------------------------------------------------------------------------------------------------------
  IMPLICIT NONE
! # Definindo precisão para os possíveis tipos de Reais
  INTEGER, PARAMETER :: ps  = SELECTED_REAL_KIND(5,30)                     ! Precisão simples
  INTEGER, PARAMETER :: pd  = SELECTED_REAL_KIND(8,100)                    ! Precisão dupla
  INTEGER, PARAMETER :: pq  = SELECTED_REAL_KIND(18,400)                   ! Precisão Quadrupla
!------------------------------------------------------------------------------------------------------------------
  INTEGER, PARAMETER :: pr  = pd                                           ! Definindo precisão padrão dos reais
!------------------------------------------------------------------------------------------------------------------
! # Definindo precisão para os possíveis tipos de inteiros
  INTEGER, PARAMETER :: ip1 = SELECTED_INT_KIND(2)
  INTEGER, PARAMETER :: ip2 = SELECTED_INT_KIND(4)
  INTEGER, PARAMETER :: ip3 = SELECTED_INT_KIND(8)
  INTEGER, PARAMETER :: ip4 = SELECTED_INT_KIND(10)
  INTEGER, PARAMETER :: ip5 = SELECTED_INT_KIND(20)
!------------------------------------------------------------------------------------------------------------------
END MODULE precisao
```

```python
%%writefile src/leitura.f95
MODULE leitura
!------------------------------------------------------------------------------------------------------------------
USE precisao
!------------------------------------------------------------------------------------------------------------------
IMPLICIT NONE
  REAL(ps),SAVE             :: t0,tf          ! Tempo incial e tempo final
  REAL(ps),SAVE             :: o10,o20        ! Ângulos inciais
  REAL(ps),SAVE             :: l10,l20        ! Posições inciais
  REAL(ps),SAVE             :: v10,v20        ! Velocidades inciais
  REAL(ps),SAVE             :: w10,w20        ! Velocidade angulares inciais
  REAL(ps),SAVE             :: l1,l2          ! Comprimento das hastes
  REAL(ps),SAVE             :: k1,k2          ! Constantes elásticas        
  REAL(ps),SAVE             :: m1,m2          ! Massa dos pêndulos 
  REAL(ps),SAVE             :: g              ! Aceleração da gravidade
  REAL(ps),SAVE             :: ut,um,uc
  INTEGER(ip3),SAVE         :: n              ! Número de iterações do método RK4
  INTEGER(ip2),SAVE         :: gl             ! Graus de liberdade do sistema (número de eq. dif. de 2º ordem)
  CHARACTER(LEN=20),SAVE    :: nome
!------------------------------------------------------------------------------------------------------------------
CONTAINS
!------------------------------------------------------------------------------------------------------------------
SUBROUTINE lercaso(caso,arq)
 IMPLICIT NONE
   INTEGER(ip1),INTENT(IN)       :: caso
   CHARACTER(LEN=20),INTENT(IN)  :: arq
OPEN(UNIT=20,FILE=arq,FORM='FORMATTED',STATUS='OLD')
    READ(20,*) t0
    READ(20,*) tf
    READ(20,*) o10
    READ(20,*) w10
    READ(20,*) o20
    READ(20,*) w20
    READ(20,*) l1
    READ(20,*) l2
    READ(20,*) m1
    READ(20,*) m2
    READ(20,*) g
    READ(20,*) n
    READ(20,*) gl
    READ(20,*) nome
    READ(20,*) ut
    READ(20,*) um
    READ(20,*) uc
! # Adimensionalização
t0  = t0/ut
tf  = tf/ut
w10 = w10*ut
w20 = w20*ut
l1  = l1/uc
l2  = l2/uc
m1  = m1/um
m2  = m2/um
g   = (g*ut*ut)/um
Select case(caso)
  case(1)
      Call ler1(arq)
  case(2)
      Call ler2(arq)
  case(3)
      Call ler3(arq)
  case(4)
      Call ler4(arq)
end select
!------------------------------------------------------------------------------------------------------------------   
END SUBROUTINE lercaso
!------------------------------------------------------------------------------------------------------------------
SUBROUTINE ler1(caso)
 IMPLICIT NONE
  CHARACTER (LEN=20),INTENT(IN) :: caso  
CLOSE(UNIT=20,STATUS='KEEP')
!------------------------------------------------------------------------------------------------------------------        
    PRINT*, ' '
    PRINT*, 'Leitura feita com sucesso. Os valores lidos são:'
    PRINT*, ' '
    PRINT*, 't0   =',t0,'s'
    PRINT*, 'tf   =',tf,'s'
    PRINT*, 'o10  =',o10,'rad'
    PRINT*, 'o20  =',o20,'rad'
    PRINT*, 'w10  =',w10,'rad/s'
    PRINT*, 'w20  =',w20,'rad/s'
    PRINT*, 'l1   =',l1,'m'
    PRINT*, 'l2   =',l2,'m'
    PRINT*, 'm1   =',m1,'Kg'
    PRINT*, 'm2   =',m2,'Kg'
    PRINT*, 'g    =',g,'m/s²'
    PRINT*, 'n    =',n
    PRINT*, 'gl   =',gl
    PRINT*, 'Nome =',nome
!------------------------------------------------------------------------------------------------------------------   
END SUBROUTINE ler1
!------------------------------------------------------------------------------------------------------------------
SUBROUTINE ler2(caso)
 IMPLICIT NONE
  CHARACTER (LEN=20),INTENT(IN)  :: caso
    READ(20,*) l10
    READ(20,*) v10
    READ(20,*) k1
! # Adimensionalização
l10  = l10/uc
v10  = (v10*ut)/uc
k1   = (k1*ut*ut)/um
CLOSE(UNIT=20,STATUS='KEEP')
!------------------------------------------------------------------------------------------------------------------        
    PRINT*, ' '
    PRINT*, 'Leitura feita com sucesso. Os valores lidos são:'
    PRINT*, ' '
    PRINT*, 't0   =',t0,'s'
    PRINT*, 'tf   =',tf,'s'
    PRINT*, 'o10  =',o10,'rad'
    PRINT*, 'o20  =',o20,'rad'
    PRINT*, 'w10  =',w10,'rad/s'
    PRINT*, 'w20  =',w20,'rad/s'
    PRINT*, 'l10  =',l10,'m'
    PRINT*, 'v10  =',v10,'m/s'
    PRINT*, 'k1   =',k1,'N/m'
    PRINT*, 'l1   =',l1,'m'
    PRINT*, 'l2   =',l2,'m'
    PRINT*, 'm1   =',m1,'Kg'
    PRINT*, 'm2   =',m2,'Kg'
    PRINT*, 'g    =',g,'m/s²'
    PRINT*, 'n    =',n
    PRINT*, 'gl   =',gl
    PRINT*, 'Nome =',nome
!------------------------------------------------------------------------------------------------------------------   
END SUBROUTINE ler2
!------------------------------------------------------------------------------------------------------------------
SUBROUTINE ler3(caso)
 IMPLICIT NONE
  CHARACTER (LEN=20),INTENT(IN) :: caso
    READ(20,*) l20
    READ(20,*) v20
    READ(20,*) k2
! # Adimensionalização
l20  = l20/uc
v20  = (v20*ut)/uc
k2   = (k2*ut*ut)/um
CLOSE(UNIT=20,STATUS='KEEP')
!------------------------------------------------------------------------------------------------------------------        
    PRINT*, ' '
    PRINT*, 'Leitura feita com sucesso. Os valores lidos são:'
    PRINT*, ' '
    PRINT*, 't0   =',t0,'s'
    PRINT*, 'tf   =',tf,'s'
    PRINT*, 'o10  =',o10,'rad'
    PRINT*, 'o20  =',o20,'rad'
    PRINT*, 'w10  =',w10,'rad/s'
    PRINT*, 'w20  =',w20,'rad/s'
    PRINT*, 'l1   =',l1,'m'
    PRINT*, 'l2   =',l2,'m'
    PRINT*, 'l20  =',l20,'m'
    PRINT*, 'v20  =',v20,'m/s'
    PRINT*, 'k2   =',k2,'N/m'
    PRINT*, 'm1   =',m1,'Kg'
    PRINT*, 'm2   =',m2,'Kg'
    PRINT*, 'g    =',g,'m/s²'
    PRINT*, 'n    =',n
    PRINT*, 'gl   =',gl
    PRINT*, 'Nome =',nome
!------------------------------------------------------------------------------------------------------------------   
END SUBROUTINE ler3
!------------------------------------------------------------------------------------------------------------------
SUBROUTINE ler4(caso)
 IMPLICIT NONE
  CHARACTER (LEN=20),INTENT(IN) :: caso    
    READ(20,*) l10
    READ(20,*) v10
    READ(20,*) k1
    READ(20,*) l20
    READ(20,*) v20
    READ(20,*) k2
! # Adimensionalização
l10  = l10/uc
v10  = (v10*ut)/uc
k1   = (k1*ut*ut)/um
l20  = l20/uc
v20  = (v20*ut)/uc
k2   = (k2*ut*ut)/um
CLOSE(UNIT=20,STATUS='KEEP')
!------------------------------------------------------------------------------------------------------------------        
    PRINT*, ' '
    PRINT*, 'Leitura feita com sucesso. Os valores lidos são:'
    PRINT*, ' '
    PRINT*, 't0   =',t0,'s'
    PRINT*, 'tf   =',tf,'s'
    PRINT*, 'o10  =',o10,'rad'
    PRINT*, 'o20  =',o20,'rad'
    PRINT*, 'w10  =',w10,'rad/s'
    PRINT*, 'w20  =',w20,'rad/s'
    PRINT*, 'l1  =',l1,'m'
    PRINT*, 'l10  =',l10,'m'
    PRINT*, 'v10  =',v10,'m/s'
    PRINT*, 'k1   =',k1,'N/m'
    PRINT*, 'l2   =',l2,'m'
    PRINT*, 'l20  =',l20,'m'
    PRINT*, 'v20  =',v20,'m/s'
    PRINT*, 'k2   =',k2,'N/m'
    PRINT*, 'm1   =',m1,'Kg'
    PRINT*, 'm2   =',m2,'Kg'
    PRINT*, 'g    =',g,'m/s²'
    PRINT*, 'n    =',n
    PRINT*, 'gl   =',gl
    PRINT*, 'Nome =',nome
!------------------------------------------------------------------------------------------------------------------   
END SUBROUTINE ler4
!------------------------------------------------------------------------------------------------------------------
END MODULE leitura
```

```python
%%writefile src/p2s.f95
MODULE p2s
  USE precisao
  USE leitura
  IMPLICIT NONE
!------------------------------------------------------------------------------------------------------------------        
CONTAINS
!------------------------------------------------------------------------------------------------------------------
FUNCTION f1(u)
    IMPLICIT NONE
    REAL(pr),DIMENSION(:),INTENT(IN)     :: u
    REAL(pr),DIMENSION(size(u))         :: f1
!------------------------------------------------------------------------------------------------------------------
!#  u(1) = o1
!#  u(2) = o2
!#  u(3) = w1
!#  u(4) = w2    
!------------------------------------------------------------------------------------------------------------------
! # Equações diferenciais   
    f1(1) = u(3)                                                                                        ! Derivada primeira temporal de theta1 
    f1(2) = u(4)                                                                                        ! Derivada primeira temporal de theta2
    f1(3) = -(m2/(m1+m2))*(l2/l1)*(f1(4)*cos(u(1)-u(2))+f1(2)*f1(2)*sin(u(1)-u(2)))-(g/l1)*sin(u(1))    ! Derivada segunda temporal de theta1
    f1(4) = -(l1/l2)*(f1(3)*cos(u(1)-u(2))-f1(1)*f1(1)*sin(u(1)-u(2)))-(g/l2)*sin(u(2))                 ! Derivada segunda temporal de theta2
!------------------------------------------------------------------------------------------------------------------
  END FUNCTION f1
!------------------------------------------------------------------------------------------------------------------
FUNCTION f2(u)
    IMPLICIT NONE
    REAL(pr),DIMENSION(:),INTENT(IN)     :: u
    REAL(pr),DIMENSION(size(u))         :: f2
    REAL(pr)                             :: a2 ! Variáveis auxiliares
!------------------------------------------------------------------------------------------------------------------
!#  u(1) = o1
!#  u(2) = o2
!#  u(3) = r1
!#  u(4) = w1
!#  u(5) = w2
!#  u(6) = v1
!------------------------------------------------------------------------------------------------------------------
! # Equações diferenciais de segunda ordem
    a2  = m2/(m1+m2)  
    f2(1) = u(4)                                                                                            ! Derivada primeira temporal de theta1 
    f2(2) = u(5)                                                                                            ! Derivada primeira temporal de theta2
    f2(3) = u(6)                                                                                            ! Derivada primeira temporal de l1
!
    f2(4) = -(2.0/(l10+u(3)))*u(6)*u(4)  -a2*(l2/(l10+u(3)))*(f2(5)*cos(u(1)& 
            -u(2))+f2(2)*f2(2)*sin(u(1)-u(2)))-(g/(l10+u(3)))*sin(u(1))                                                               ! Derivada segunda temporal de theta1
!
    f2(5) = -(1.0/l2)*f2(6)*sin(u(1)-u(2))-(2.0/l2)*u(6)*u(4)*cos(u(1)-u(2))&
    -(l10+u(3)/l2)*(f2(4)*cos(u(1)-u(2))-u(4)*u(4)*sin(u(1)-u(2)))-(g/l2)*sin(u(2))
!
    f2(6) = -a2*l2*(f2(5)*sin(u(1)-u(2))-u(5)*u(5)*cos(u(1)-u(2))) & 
            +(l10+u(3))*u(4)*u(4)+g*cos(u(1))-(k1/(m1+m2))*(l10+u(3)-l1)
!------------------------------------------------------------------------------------------------------------------
  END FUNCTION f2
!------------------------------------------------------------------------------------------------------------------
FUNCTION f3(u3)
    IMPLICIT NONE
    REAL(pr),DIMENSION(:),INTENT(IN)     :: u3
    REAL(pr),DIMENSION(size(u3))         :: f3
    REAL(pr)                             :: a3
!------------------------------------------------------------------------------------------------------------------
!#  u3(1) = o1
!#  u3(2) = o2
!#  u3(3) = r2
!#  u3(4) = w1 
!#  u3(5) = w1 
!#  u3(6) = v2 
!------------------------------------------------------------------------------------------------------------------
! # Equações diferenciais   
a3  = m2/(m2+m2)
    f3(1) = u3(4)                                                                                               ! Derivada primeira temporal de theta1 
    f3(2) = u3(5)                                                                                               ! Derivada primeira temporal de theta2
    f3(3) = u3(6)                                                                                               ! Derivada primeira temporal de l2
!    
    f3(4) = -a3*((l20+u3(3))/l1)*(f3(5)*cos(u3(1)-u3(2))+f3(2)*f3(2)*sin&
    (u3(1)-u3(2)))+a3*(1/l1)*(f3(6)*sin(u3(1)-u3(2))+f3(3)*f3(3)*cos(u3(1)-u3(2)))-(g/l1)*sin(u3(1))
!
    f3(5) = (l1/(l20+u3(3)))*(f3(4)*cos(u3(1)-u3(2))-f3(1)*f3(1)*sin(u3(1)-u3(2)))-(g/(l20+u3(2)))*sin(u3(1)-u3(2))
!
    f3(6) = l1*(f3(4)*sin(u3(1)-u3(2))-f3(1)*f3(1)*cos(u3(1)-u3(2)))+(l20+u3(4))*f3(2)*f3(2)-g*cos(u3(2))-(k2/m2)*(l20+u3(4)-l2)
!------------------------------------------------------------------------------------------------------------------
  END FUNCTION f3
!------------------------------------------------------------------------------------------------------------------
FUNCTION f4(u4)
    IMPLICIT NONE
    REAL(pr),DIMENSION(:),INTENT(IN)     :: u4
    REAL(pr),DIMENSION(size(u4))         :: f4
!------------------------------------------------------------------------------------------------------------------
!#  u4(1) = o1
!#  u4(2) = o2
!#  u4(3) = r1
!#  u4(4) = r2 
!#  u4(5) = w1
!#  u4(6) = w2
!#  u4(7) = v1
!#  u4(8) = v2 
!------------------------------------------------------------------------------------------------------------------
! # Equações diferenciais    
    f4(1) = u4(5)                                                                                            ! Derivada primeira temporal de theta1 
    f4(2) = u4(6)                                                                                            ! Derivada primeira temporal de theta2
    f4(3) = u4(7)                                                                                            ! Derivada primeira temporal de l1
    f4(4) = u4(8)                                                                                            ! Derivada primeira temporal de l2
!    
    f4(5) = -(m2/(m1+m2))*(l20+u4(4)/(l10+u4(3)))*f4(6)*cos(u4(1)&
    -u4(2))-(m2/(m1+m2))*(1/(l10+u4(3)))*(f4(8)*sin(u4(1)-u4(2))+&
    f4(4)*f4(2)*cos(u4(1)-u4(2))+f4(1)*f4(3)*cos(u4(1)-u4(2)))+2*&
    (m2/(m1+m2))*(1/((l10+u4(3))*(l10+u4(3))))*f4(3)*f4(4)*cos(u4(1))*&
    cos(u4(2))+(g/(l10+u4(3)))*sin(u4(1))     ! Derivada segunda temporal de theta1
!    
    f4(6) = -(((l10+u4(3))/(l20+u4(4)))*(f4(1)*cos(u4(1)-u4(2))&
    -f4(1)*f4(1)*sin(u4(1)-u4(2)))-(1/(l20+u4(4)))*f4(7)*sin(u4(1)&
    -u4(2))-(2/(l20+u4(4)))*f4(3)*f4(1)*cos(u4(1)-u4(2))-(2/(l20+&
    u4(4))*(l20+u4(4)))*f4(3)*f4(4)*cos(u4(1)-u4(2))*sin(u4(1)-u4(2))-(g/(l20+u4(4)))&
    *sin(u4(2)))         ! Derivada segunda temporal de theta2
!    
    f4(7) =  u4(4)    ! Será escrito aqui o resto do sistema de equações diferenciais 
!    
    f4(8) = u4(4)     ! Será escrito aqui o resto do sistema de equações diferenciais
!------------------------------------------------------------------------------------------------------------------
  END FUNCTION f4
!------------------------------------------------------------------------------------------------------------------
END MODULE p2s
```
